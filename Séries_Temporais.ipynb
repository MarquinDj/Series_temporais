```python
def plotar(titulo, labelx, labely, x, y, dataset):

  sns.set_palette('Accent')
  sns.set_style('darkgrid')
  ax = sns.lineplot(x=x, y=y, data=dataset)
  ax.figure.set_size_inches(12,6)
  ax.set_title(titulo, loc='left', fontsize = 18)
  ax.set_xlabel(labelx, fontsize =14)
  ax.set_ylabel(labely, fontsize =14)
  ax = ax
```

```python
plotar('Aumento das vendas da Alucar de 2017 a 2018', 'Tempo', 'Aumento', 
       'mes', 'aumento', alucar)
```

```python
alucar['aceleracao'] = alucar['aumento'].diff()
alucar.head()
```

```python
plotar('AceleraÃ§Ã£o das vendas da Alucar de 2017 a 2018', 'Tempo', 'AceleraÃ§Ã£o', 
       'mes', 'aceleracao', alucar)
```

```python
plt.figure(figsize=(16,12))
ax = plt.subplot(3,1,1)
ax.set_title('AnÃ¡lise de Vendas da alucar de 2017 a 2018', fontsize = 18, loc='left')
sns.lineplot(x='mes', y='vendas', data=alucar)
plt.subplot(3, 1,2)
sns.lineplot(x='mes', y='aumento', data=alucar)
plt.subplot(3, 1,3)
sns.lineplot(x='mes', y='aceleracao', data=alucar)
ax = ax
```

```python
def plot_comparacao(x, y1, y2, y3, dataset, titulo): 
  plt.figure(figsize=(16,12))
  ax = plt.subplot(3,1,1)
  ax.set_title(titulo, fontsize = 18, loc='left')
  sns.lineplot(x=x, y=y1, data=dataset)
  plt.subplot(3, 1,2)
  sns.lineplot(x=x, y=y2, data=dataset)
  plt.subplot(3, 1,3)
  sns.lineplot(x=x, y=y3, data=dataset)
  ax = ax
```

```python
plot_comparacao('mes', 'vendas', 'aumento', 'aceleracao', 
                alucar, 'AnÃ¡lise de Vendas da alucar de 2017 a 2018')
```

```python
from pandas.plotting import autocorrelation_plot 
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('CorrelaÃ§Ã£o das vendas', fontsize = 18, x = 0.26, y = 0.95)
autocorrelation_plot(alucar['vendas'])
ax=ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('CorrelaÃ§Ã£o do aumento', fontsize = 18, x = 0.26, y = 0.95)
autocorrelation_plot(alucar['aumento'][1:])
ax=ax
```

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle('CorrelaÃ§Ã£o da aceleracao', fontsize = 18, x = 0.26, y = 0.95)
autocorrelation_plot(alucar['aceleracao'][2:])
ax=ax
```

**Analisando a Sazonalidade** 

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia'].dt.day_name()
```

```python
vendas_por_dia.head()
```

```python
vendas_por_dia['dia_da_semana'].unique()
```

```python
dias_traduzidos = {'Monday': 'Segunda', 'Tuesday': 'TerÃ§a', 'Wednesday':'Quarta', 
                    'Thursday':'Quinta', 'Friday':'Sexta', 'Saturday':'Sabado',
       'Sunday':'Domingo'}
```

```python
vendas_por_dia['dia_da_semana'] = vendas_por_dia['dia_da_semana'].map(dias_traduzidos)
```

```python
vendas_por_dia.head()
```

**Agrupando os Dias**

```python
vendas_agrupadas = vendas_por_dia.groupby('dia_da_semana')['vendas', 'aumento', 
'aceleracao'].mean().round()
```

```python
vendas_agrupadas
```

**CorrelaÃ§Ã£o das vendas DiÃ¡rias**

```python
ax = plt.figure(figsize = (12,6))
ax.suptitle('COrrelaÃ§Ã£o das Vendas diÃ¡rias', fontsize = 18, x = 0.3, y = 0.95)
autocorrelation_plot(vendas_por_dia['vendas'])
ax= ax
```

```python
ax = plt.figure(figsize = (12,6))
ax.suptitle('COrrelaÃ§Ã£o do Aumento das Vendas diÃ¡rias', fontsize = 18, x = 0.3, y = 0.95)
autocorrelation_plot(vendas_por_dia['aumento'][1:])
ax= ax
```

```python
ax = plt.figure(figsize = (12,6))
ax.suptitle('COrrelaÃ§Ã£o da AceleraÃ§Ã£o das Vendas diÃ¡rias', fontsize = 18, x = 0.3, y = 0.95)
autocorrelation_plot(vendas_por_dia['aceleracao'][2:])
ax= ax
```

# Statsmodels

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose([chocolura['vendas']], freq = 3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
data = ({
    'observacao': observacao,
    'tendencia': tendencia,
    'sazonalidade': sazonalidade,
    'ruido': ruido
})

resultado = pd.DataFrame(data)
resultado.head()
```

```python
plot_comparacao(resultado.index, 'observacao', 'tendencia', 'sazonalidade', resultado, 'Exemplo de StatsModels')
```
